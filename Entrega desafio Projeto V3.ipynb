```python
import requests
import schedule
import time
import pandas as  pd
from plyer import notification 
from datetime import datetime
from colorama import Fore, Back, Style
```

```python
'''Tenho uma lista com os 4 URLs para conectar a APIs, onde a URL https://dummy.restapiexample.com/api/v1/employees da error 406, 
as outras tem sucesso na conexão'''

url = ['https://restcountries.com/v3.1/all','https://brasilapi.com.br/api/banks/v1','https://dummy.restapiexample.com/api/v1/employees','https://jsonplaceholder.typicode.com/posts']
```

```python
#Funcão de alerta para validar se a conexão teve sucesso
def alerta(code, url, etapa):

    now = (datetime.now())
    data_hora = now.strftime("%d/%m/%y %H:%M:%S")
    msg = f'Falha no carregamento da base {url} na etapa {etapa}. \n{data_hora}'
    

    if code == 200:
        title = 'Conexão exitosa'
        msg = f'Voce esta conectado na base {url}. \n{data_hora}'
    elif code >= 400:
        title = 'Não tem permissão de acessar na base: {url}'
    elif code >= 500:
        title = 'Erro de conexão com o server {url}'

    notification.notify(
        title = title,
        message = msg,
        app_name= 'Alerta',
        timeout = 10
    )
    
#criando uma lista vazia para almazenar cada request dos APIs com código 200
list=[]

#criando um for para iterar cada url da lista e criar a conexão, se o código é 200, adiciona essa conexão na lista list
for url in url:
    req = requests.get(url)
    code = req.status_code
    if code == 200:
        list.append(req.json()) #adiciono cada conexão na ista
    
    alerta(code, url, etapa = 'Extracao')
    print(url, code)
list[1] #para testar a lista list
```

```python
'''Aqui se cria cada data frame para as 3 APIs que tiveram sucesso na conexão'''

df = pd.DataFrame(list[0])
df2 = pd.DataFrame(list[1])
df3 = pd.DataFrame(list[2])
df.head(4)
```
